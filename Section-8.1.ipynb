```julia
using Primes
using Printf
```

```julia
function powers(n)
    @printf("   ")
    for k in 1:totient(n)
        @printf("%3u", k)
    end
    println()
    A = [a for a in 2:(n-1) if gcd(a,n)==1]
    for a::BigInt in A
        @printf("%3u", a)
        for k in 1:totient(n)
            @printf("%3u", (a^k)%n)
        end
        println()
    end
end
```
