```julia
using Printf

for n in 1:12
    @printf("%2u & %11s & %2u & %11s & %2u & %11s \\\\\n", n, to_ppn(n), n+12, to_ppn(n+12), n+24, to_ppn(n+24))
end
```

```julia
for n in 1:1023
    x = to_ppn(n)
    length(x) == 5 && @printf("%8s\t%4u\n", to_ppn(n), n)
end
```
